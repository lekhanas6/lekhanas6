### Data Analysis

#### Hypothetical Case Study: Skyline Taxis

**Background:**
Skyline Taxis is a prominent taxi company operating in New York City. With the rise of ride-sharing apps, the traditional taxi business has faced challenges. Skyline Taxis wants to remain competitive and is considering a revamp of its operations based on insights drawn from its data.

**Objective:**
The company aims to optimize its fleet operations, improve rider experience, and target marketing efforts more effectively.

1. **What are the peak hours for taxi rides?**
    - **Importance:** Peak hours indicate when the demand is highest. Knowing this can help optimize driver schedules and taxi availability.

2. **Which hours of the day have the lowest average fare amount and the highest average fare amount?**
    - **Importance:** Identifying these hours can help in setting offering promotions during low-fare hours to boost demand.
  
    
3. **What are the peak times for airport pickups?**
    - **Importance:** Airport rides are generally longer and more profitable. Identifying peak times can help in allocating more taxis to the airport during these hours.
    
    
4. **What is the distribution of trip distance?**
    - **Importance:** This can help in understanding the nature of most rides, guiding decisions on vehicle type (small cars for shorter trips, comfortable cars for longer trips).
    







#### 1) What are the peak hours for taxi rides?
* This information can be valuable for taxi service providers as they can allocate more resources during peak hours to meet the demand.

```python
# First, we want to extract the hour from the pickup datetime.
# The .dt.hour method helps us easily extract the hour from a datetime column.
data['pickup_hour'] = data['tpep_pickup_datetime'].dt.hour
```

```python
data.head()
```

```python
# Let's take a look at the head of our dataset again
```

```python
# Now, we want to know:
# for each hour of the day, from 0 (midnight) to 23 (11 PM), how many taxi pickups occurred?
rush_hours = data['pickup_hour'].value_counts().reset_index()
```

```python
rush_hours
```

```python
# Let's save this into a new variable called rush_hours
```

```python
# And change the column names to hour and count
rush_hours.columns = ['hour', 'count']
```

```python
# Let's take a look at the final dataset
rush_hours
```

```python
# Let's sort this dataset by hour
rush_hours = rush_hours.sort_values(by='hour')
```

```python
rush_hours
```

```python
# Let's visualize this dataset
px.bar(rush_hours, x='hour', y='count', title='taxi ride count by hour', text='count')
```

#### 2) Which hours of the day have the lowest average fare amount and the highest average fare amount?

* Offer time-based promotions during these off-peak hours to incentivize rides and increase demand.
* Promote scheduled rides or advance bookings during the peak hours at a reduced rate.
* Reduce the number of drivers or vehicles on the road during off hours to save on operational costs.



```python
# Let's group the data by the hour we just extracted.
# Our goal is to find out the average fare amount for each hour.
off_peak_hours = data.groupby('pickup_hour')['fare_amount'].mean().reset_index()
```

```python
# Let's reset index
```

```python
# Let's take a look at the data
off_peak_hours
```

```python
# round to two decimal points: to increase readability
off_peak_hours['fare_amount'] = round(off_peak_hours['fare_amount'], 2)
```

```python
off_peak_hours
```

```python
# Now, let's visualize our findings.
px.bar(off_peak_hours, x='pickup_hour', y='fare_amount')
```

#### 3) What are the peak times for airport pickups?


* Taxi companies or ride-sharing platforms can allocate more vehicles or drivers to the airport during peak times, ensuring that there's always a vehicle available for passengers. This can lead to increased rides and revenue.


```python
# Let's create a subset of our main dataset, specifically focusing on airport rides.
airport_rides = data[data['airport_fee'] != 0]
```

```python
airport_rides
```

```python
# Let's count the number of airport rides that occurred during each hour.
airport_pickups = airport_rides.groupby('pickup_hour').size().reset_index()
```

```python
# Let' reset the index
airport_pickups
```

```python
# change the column names
airport_pickups.columns = ['pickup_hour', 'count']
```

```python
airport_pickups
```

```python
# Visualize
px.bar(airport_pickups, x='pickup_hour', y='count')
```
